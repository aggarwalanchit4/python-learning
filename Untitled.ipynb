```python
class FractionConverter:
    def __init__(self, num):
        self.num = float(num)

    def gcd(self, a, b):
        while b:
            a, b = b, a % b
        return a

    def convert(self):
        s = str(self.num)
        if '.' not in s:
            return f"{int(self.num)}/1"

        decimal_places = len(s.split('.')[1])
        denominator = 10 ** decimal_places
        numerator = int(self.num * denominator)

        g = self.gcd(numerator, denominator)

        return f"{numerator // g}/{denominator // g}"

num = input()
obj = FractionConverter(num)

print(obj.convert())
```
